```python
from datetime import timedelta, date

startDate = date(2018,8,27)
endDate = date(2018,12,10)

# calculate how many days are there from start to end of semester
day_count = (endDate - startDate).days + 1

current_week = 0
start_weekday = startDate.weekday()

# this needs to be set up manually
university_holiday = [[date(2018,9,3),'Labor Day'],
                      [date(2018,9,4),'Last Day of Add/Drop'],
                      [date(2018,9,5),'First Day of Course Withdrawal'],
                      [date(2018,10,23),'Last Day of Course Withdrawal'],
                      [date(2018,11,19),'Reading/Writing Day'],
                      [date(2018,11,20),'Reading/Writing Day'],
                      [date(2018,11,21),'Fall/Thanksgiving Break'],
                      [date(2018,11,22),'Fall/Thanksgiving Break'],
                      [date(2018,11,23),'Fall/Thanksgiving Break'],
                      [date(2018,11,24),'Fall/Thanksgiving Break'],
                      [date(2018,11,25),'Fall/Thanksgiving Break'],
                      [date(2018,12,10),'Last Day of the Session'],
                      [date(2018,12,11),'First Day of Final Exams'],
                      [date(2018,12,12),'Last Day of Final Exams']]


current_holiday = 0
for i in range(0,day_count):
    current_date = startDate + timedelta(i)
    isHoliday = False
    if current_date == university_holiday[current_holiday][0]:
        isHoliday = True
    
    #if current_date.weekday() in [1,3]:
    if current_date.weekday() in [0,2,4]:
        if current_date.weekday() == start_weekday:
            current_week += 1
            print ('\n#### Week', current_week)
        print(current_date.strftime('- %a, %b %d, %Y'))
        if isHoliday:
            print(university_holiday[current_holiday][1])
    
    if isHoliday:
        current_holiday += 1
```
